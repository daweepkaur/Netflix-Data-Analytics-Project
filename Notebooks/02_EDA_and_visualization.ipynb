```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('../data/netflix_cleaned.csv')

# 1. Movie vs TV Shows
sns.countplot(x='type', data=df, palette='Set2')
plt.title('Movies vs TV Shows')
plt.show()

# Top 10 Countries (cleaned)
df_country = df.dropna(subset=['country'])
df_country = df_country[df_country['country'].str.lower() != 'unknown']

country_counts = df_country['country'].value_counts().head(10)

plt.figure(figsize=(8,4))
country_counts.plot(kind='bar', color='skyblue')
plt.title('Top 10 Countries Producing Netflix Content')
plt.ylabel('Number of Titles')
plt.xticks(rotation=45)
plt.show()

# 3. Titles added per year
sns.histplot(df['year_added'], bins=20, color='purple')
plt.title('Titles Added per Year')
plt.show()
```
