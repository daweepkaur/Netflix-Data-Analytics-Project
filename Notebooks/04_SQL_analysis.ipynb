```python
import sqlite3
import pandas as pd

df = pd.read_csv('../data/netflix_enriched.csv')
conn = sqlite3.connect('../data/netflix.db')
df.to_sql('netflix', conn, if_exists='replace', index=False)

# Run SQL queries
query = "SELECT type, COUNT(*) AS total FROM netflix GROUP BY type;"
pd.read_sql(query, conn)
```
